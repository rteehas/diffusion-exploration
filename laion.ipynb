```python
import ast
import bisect
counts = []

sorted_pairs = []
with open("seedselect_data/unigram_frequencies.txt", "r", encoding="utf-8") as f:
    for raw in f:
        # drop the trailing comma, turn the tuple string into a tuple object
        word, n = ast.literal_eval(raw.rstrip(',\n').strip("[").strip("]"))
        counts.append((word, n))

        # bisect.insort(sorted_pairs, (n, word))
```

```python
sorted_words = sorted(counts, key=lambda t: t[1])
```

```python
sorted_words[-1]
```

```python
l_bound = 5000
u_bound = 6000

selected = [c for c in sorted_words if l_bound <= c[1] and c[1] <= u_bound]
```

```python
import numpy as np

sample = np.random.choice(len(selected), size=10, replace=False).tolist()
sample = [selected[i] for i in sample]
```

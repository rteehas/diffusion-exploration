```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
no_tail_true = pd.read_csv("wandb_csvs/wandb_export_2025-08-01T13_52_32.740-04_00.csv")
no_tail_false = pd.read_csv("wandb_csvs/wandb_export_2025-08-01T13_53_00.275-04_00_false.csv")
```

```python
no_tail_true
```

```python
no_tail_false
```

```python
offwhite = "#F7F6F2"
plt.figure(facecolor="#F7F6F2")

fig, ax = plt.subplots()
fig.patch.set_facecolor("#F7F6F2")
ax.plot(no_tail_true["Step"], no_tail_true["ddpo_trainer_config.train_learning_rate: 0.00005 - head_validation/reward_mean"], label="Top 10% Classes Only", color="#203A63")
ax.plot(no_tail_false["Step"], no_tail_false["ddpo_trainer_config.train_learning_rate: 0.00005 - head_validation/reward_mean"], label="Top 10% + Bottom 10%", color="#C94C4C")
ax.legend(frameon=False)
plt.title("ImageReward - ImageNet Top 10%")
ax.set_facecolor(offwhite)
ax.tick_params(colors="#333333")
ax.xaxis.label.set_color("#333333")
ax.yaxis.label.set_color("#333333")
ax.set_xlabel("Step")
ax.set_ylabel("Reward")
```
